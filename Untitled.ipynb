```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt
from sklearn.decomposition import FastICA
from scipy.stats import kurtosis

from scipy.signal import welch
from scipy.stats import skew
```

```python
data = pd.read_csv("sample.csv")
data.head()
```

```python
# selected the features
base_channels = ['F7', 'AF3', 'F3', 'F4', 'AF4', 'F8', 'T7', 'T8']

# Ensure the dataset contains a Timestamp column
if 'Timestamp' not in data.columns:
    raise ValueError("The dataset does not contain a 'Timestamp' column.")

# Build a list of relevant columns
relevant_columns = ['Timestamp']  # Start with the timestamp
for channel in base_channels:
    relevant_columns.extend(
        [col for col in data.columns if channel in col]
    )

# Extract the filtered data
filtered_data = data[relevant_columns]

# Display the first few rows
print(filtered_data.head())
```

```python
#saving the extracted data to a csv file 
#filtered_data.to_csv("splitted_channels.csv", index=False)
splitted_channels=pd.read_csv("splitted_channels.csv")
```

```python
#replaced my null values with the median
splitted_channels=splitted_channels.fillna(splitted_channels.median())
```

```python
# Assuming your dataset has a Timestamp column and EEG signal columns
if 'Timestamp' not in splitted_channels.columns:
    raise ValueError("The dataset does not contain a 'Timestamp' column.")

# Ensure Timestamp is in datetime format
splitted_channels['Timestamp'] = pd.to_datetime(splitted_channels['Timestamp'])

# Function to calculate sampling frequency for a single column
def calculate_sampling_frequency(timestamp_col):
    time_differences = timestamp_col.diff().dt.total_seconds()  # Calculate time differences
    average_time_interval = time_differences.dropna().mean()  # Average time interval in seconds
    return 1 / average_time_interval if average_time_interval else None  # Sampling frequency in Hz

# Create a dictionary to store the sampling frequency for each column
sampling_frequencies = {}

# Loop through EEG signal columns
eeg_columns = [col for col in splitted_channels.columns if col.startswith("EEG")]
for col in eeg_columns:
    # Calculate sampling frequency for the current column using its timestamps
    sampling_frequencies[col] = calculate_sampling_frequency(splitted_channels['Timestamp'])

# Display the sampling frequencies
for col, freq in sampling_frequencies.items():
    print(f"Sampling Frequency for {col}: {freq:.2f} Hz")
```

```python

# Define bandpass filter
def bandpass_filter(signal, lowcut, highcut, fs, order=5):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, signal)
# Ensure Timestamp is in datetime format
if 'Timestamp' not in splitted_channels.columns:
    raise ValueError("The dataset does not contain a 'Timestamp' column.")

try:
    splitted_channels['Timestamp'] = pd.to_datetime(splitted_channels['Timestamp'])
except Exception as e:
    raise ValueError(f"Failed to convert 'Timestamp' to datetime. Error: {e}")


# Parameters for filtering
lowcut = 0.5  # Lower bound frequency in Hz
highcut = 50.0  # Upper bound frequency in Hz


# Apply bandpass filter to each signal column
eeg_columns = [col for col in splitted_channels.columns if col.startswith("EEG")]
filtered_data = splitted_channels.copy()
for col in eeg_columns:
    filtered_data[col] = bandpass_filter(splitted_channels[col].fillna(0), lowcut, highcut, 128)
    
# Display the first few rows of the filtered data
print(filtered_data.head())
```

```python
def calculate_snr(signal, noise):
    signal_power = np.mean(signal ** 2)
    noise_power = np.mean(noise ** 2)
    if noise_power == 0:  # Avoid division by zero
        return np.inf
    snr = 10 * np.log10(signal_power / noise_power)
    return snr
```

```python
snr_results = {}
for channel in eeg_columns:
    # Original signal
    signal = filtered_data[channel].values
    
    # Noise estimation: Use the original (unfiltered) signal outside the bandpass range
    noise = splitted_channels[channel].values - signal
    
    # Calculate SNR
    snr_results[channel] = calculate_snr(signal, noise)

# Display SNR for each channel
print("SNR (in dB) for each EEG channel:")
for channel, snr_value in snr_results.items():
    print(f"{channel}: {snr_value:.2f} dB")
```

```python
if 'Timestamp' not in filtered_data.columns:
    raise ValueError("The dataset does not contain a 'Timestamp' column.")

# Ensure Timestamp is in datetime format
filtered_data['Timestamp'] = pd.to_datetime(filtered_data['Timestamp'])

# Function for baseline correction
def baseline_correction(signal, baseline_range=None):
    """
    Perform baseline correction by subtracting the mean of the signal.
    """
    if baseline_range is not None:
        baseline_start, baseline_end = baseline_range
        baseline_mask = (filtered_data['Timestamp'] >= baseline_start) & (filtered_data['Timestamp'] <= baseline_end)
        baseline_mean = signal[baseline_mask].mean()
    else:
        baseline_mean = signal.mean()

    return signal - baseline_mean

# Apply baseline correction to all EEG channels
baseline_corrected_data = filtered_data.copy()
for col in eeg_columns:
    baseline_corrected_data[col] = baseline_correction(filtered_data[col])
```

```python
snr_results = {}
for channel in eeg_columns:
    # Original signal
    signal = baseline_corrected_data[channel].values
    
    # Noise estimation: Use the original (unfiltered) signal outside the bandpass range
    noise = filtered_data[channel].values - signal
    
    # Calculate SNR
    snr_results[channel] = calculate_snr(signal, noise)

# Display SNR for each channel
print("SNR (in dB) for each EEG channel:")
for channel, snr_value in snr_results.items():
    print(f"{channel}: {snr_value:.2f} dB")
```

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import FastICA

# Assuming baseline_corrected_data is your DataFrame
eeg_data_numeric = baseline_corrected_data.drop(columns=['Timestamp'])
eeg_data = eeg_data_numeric.values  # Convert to NumPy array for ICA

# Sampling frequency (adjust according to your data)
fs = 128  # Hz

# Apply ICA
ica = FastICA(n_components=eeg_data_numeric.shape[1], random_state=42, max_iter=1000)
ica_components = ica.fit_transform(eeg_data)  # Independent components

# Reconstruct the original data correctly using ICA components and mixing matrix
reconstructed_data = ica_components.dot(ica.mixing_.T)

# Check if the original data and reconstructed data have the same number of columns
if eeg_data.shape[1] != reconstructed_data.shape[1]:
    raise ValueError(f"Original data columns: {eeg_data.shape[1]}, Reconstructed data columns: {reconstructed_data.shape[1]}")

# Calculate noise (difference between the original data and the reconstructed data)
noise = eeg_data - reconstructed_data

# Calculate Noise Power (Variance of the noise)
noise_power = np.var(noise, axis=0)

# Calculate Signal Power (Variance of the ICA components)
signal_power = np.var(ica_components, axis=0)

# Calculate SNR for each component
snr = signal_power / noise_power

# Create a DataFrame for SNR results with the same column names as original EEG data
snr_df = pd.DataFrame(snr, index=eeg_data_numeric.columns, columns=['SNR'])
pd.options.display.float_format = '{:.6f}'.format  # Set the display format for floating point numbers

# Output the SNR for each component
print(snr_df)
```

```python
ica_df = pd.DataFrame(ica_components, columns=eeg_data_numeric.columns)
# Calculate kurtosis for each ICA component
kurtosis_values = np.apply_along_axis(kurtosis, axis=0, arr=ica_components)

# Create a DataFrame to store kurtosis values with the same column names as the ICA components
kurtosis_df = pd.DataFrame(kurtosis_values, index=eeg_data_numeric.columns, columns=['Kurtosis'])

# Display the kurtosis values for each component
print(kurtosis_df)
```

```python
from scipy.signal import welch
import numpy as np

def extract_features(eeg_data, fs):
    """
    Extract time-domain and frequency-domain features.
    :param eeg_data: EEG data as a pandas DataFrame (channels as columns).
    :param fs: Sampling frequency.
    :return: A dictionary with extracted features.
    """
    features = {}

    for col in eeg_data.columns:
        signal = eeg_data[col]

        # Time Domain Features
        features[f'{col}_mean'] = np.mean(signal)
        features[f'{col}_var'] = np.var(signal)
        features[f'{col}_kurtosis'] = kurtosis(signal)
        features[f'{col}_skew'] = np.mean((signal - np.mean(signal))**3) / (np.std(signal)**3)

        # Frequency Domain Features (Power Spectral Density)
        f, psd = welch(signal, fs=fs, nperseg=fs*2)
        features[f'{col}_delta'] = np.sum(psd[(f >= 0.5) & (f < 4)])
        features[f'{col}_theta'] = np.sum(psd[(f >= 4) & (f < 8)])
        features[f'{col}_alpha'] = np.sum(psd[(f >= 8) & (f < 13)])
        features[f'{col}_beta'] = np.sum(psd[(f >= 13) & (f < 30)])
        features[f'{col}_gamma'] = np.sum(psd[(f >= 30) & (f < 50)])

    return features

# Example Usage
fs = 128  # Sampling frequency
features = extract_features(ica_df[eeg_columns], fs)
print("Extracted Features:", features)
```

```python

#Standardization (Z-score Normalization)
#Standardization transforms data to have a mean of 0 and a standard deviation of 1.
def standardize_signal(data):
    standardized_data = (data - data.mean()) / data.std()
    return standardized_data

# Apply standardization across channels
standardized_eeg = ica_df.apply(standardize_signal, axis=0)

print("Standardized EEG Data:")
print(standardized_eeg)
```
